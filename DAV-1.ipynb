```python

plt.hist(ds['Confirmed'], bins=10, color='skyblue', edgecolor='black')

plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')
plt.title('Histogram of Confirmed COVID-19 Cases')

plt.show()
```

```python
subset_ds= ds.head(5)


sns.scatterplot(x='Country/Region', y='Confirmed', data=subset_ds)


plt.title('Scatter Plot')

plt.show()
```

```python

subset_ds = ds.head(5)

labels = subset_ds['Country/Region']
sizes = subset_ds['Confirmed']

# Create a pie chart
plt.figure(figsize=(8, 8))  
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

plt.title('Sample Pie Chart')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = subset_ds['Country/Region']
y1 = subset_ds['Confirmed']
y2 = subset_ds['Active']
y3 = subset_ds['Deaths']

# Create multiple line graph
plt.figure(figsize=(10, 6))  # Set figure size

# Plot lines
plt.plot(x, y1, marker='o', label='Confirmed')
plt.plot(x, y2, marker='s', label='Active')
plt.plot(x, y3, marker='^', label='Deaths')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Total no. of Cases')
plt.title('Multiple Line Graph')


plt.legend()


plt.grid(True) 
plt.show()
```
